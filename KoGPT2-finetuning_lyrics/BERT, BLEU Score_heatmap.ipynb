### 데이터 - 데이터 간 BERT Score

```python
from pandas import Series, DataFrame

raw_data = {'IU': [0.6421,0.6214, 0.6152],
            'Sunwoojunga': [0.6214, 0.6328, 0.6114],
            'MonstaX': [0.6152, 0.6114, 0.6228]}

df_dd_bert = DataFrame(raw_data)
df_dd_bert.rename(index={0:'IU', 1:'Sunwoojunga', 2:'MonstaX'},inplace='True')
print(df_dd_bert)
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

ax=sns.heatmap(df_dd_bert, cmap='Blues', annot=True)
plt.xlabel("BERT Score")
```
