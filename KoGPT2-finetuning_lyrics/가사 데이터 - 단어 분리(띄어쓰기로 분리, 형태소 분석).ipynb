##  가사데이터 -> 단어로 분리

```python
## 띄어쓰기로 구분
from collections import Counter


# 전체 단어 넣을 리스트
all_words = []


for i in range(num_df):
    text = df['iu_lyrics'][i]
    words = text.split(' ')
    all_words.append(words)
    print(i, len(words))

print(len(all_words)) # ->  이게 REFERENCE가 되는 것
# unique_words = list(set(all_words)) # 중복 제거한 형태소 개수 -> 이게 REFERENCE가 되는 것
# print(len(unique_words))
```

```python
from nltk.translate.bleu_score import sentence_bleu
reference = all_words # 이대로 하면되고 (전체 형태소)
candidate = new_words # 새로 생성한 가사 형태소 분석해서 넣으면 됨. 
score = sentence_bleu(reference, candidate)
print('Cumulative 1-gram: %f' % sentence_bleu(reference, candidate, weights=(1, 0, 0, 0)))
print('Cumulative 2-gram: %f' % sentence_bleu(reference, candidate, weights=(0.5, 0.5, 0, 0)))
print('Cumulative 3-gram: %f' % sentence_bleu(reference, candidate, weights=(0.33, 0.33, 0.33, 0)))
print('Cumulative 4-gram: %f' % sentence_bleu(reference, candidate, weights=(0.25, 0.25, 0.25, 0.25)))
```

## 가사 데이터 -> 형태소로 분리

```python
## 형태소 토큰화
from konlpy.tag import Kkma
from collections import Counter
import itertools

# 형태소 분석 모델 생성
kkma = Kkma()

# 전체 단어 넣을 리스트
all_words = []


for i in range(num_df):
    text = df['iu_lyrics'][i]
    words = kkma.morphs(text)
    all_words.append(words)
    print(i, len(words))

print(len(all_words)) # ->  이게 REFERENCE가 되는 것
# unique_words = list(set(all_words)) # 중복 제거한 형태소 개수 -> 이게 REFERENCE가 되는 것
# print(len(unique_words))
```
