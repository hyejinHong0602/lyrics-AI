#### 나머지 컬럼 다 없애고 가사만 남기기. 컬럼명 lyrics로 바꾸기.

```python
import pandas as pd

df=pd.read_csv('./dataset_before/안예은_lyrics_clean.tsv', sep = '\t', encoding='cp949')
```

```python
df
```

```python
df.info()
```

```python
df=df[~df['제목'].str.contains("Ver.", na=False, case=False)] #정규식으로 english ver 포함된 행 삭제
```

```python
df=df.drop(['Unnamed: 0','제목','가수'], axis=1)
```

```python
df.columns=['lyrics']
```

```python
df.to_csv('./dataset/안예은_lyrics.tsv',sep='\t',index=False)
```
